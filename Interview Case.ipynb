### Dummy variable creation for kjonn

<h3 style="color: green"> Combining each two rows for each customer by 'group by'</h3>

```python
print(df.columns)
```

<h3 style="color: green"> Age is problematic, the same client got birthday on the day data collected which would produce two different clients if we grouped by alltogether. So, we have to group by 'alder' and the rest separately and then merge two tables </h3>

```python
df_grouped_1=df.groupby(['KUNDE_ID','LANGTIDSOVERVAKET_FLG','LO_MEDLEM_FLG','FAM_VOKSNE_ANT','FAM_BARN_ANT',
                       'DOD','KONKURS','HAR_EPOST','HAR_MOBIL','RESERVERT_MOT_MARKEDSFORING','RESERVERT_TELEFON_BRREG_FLG',
                       'MANN','KVINNE'])[['INSKUDD','UTLAN','LONNSOMHET_PER_MND','PURRING_ANTALL_12MND',
                                          'ANT_INNLOGGINGER_MOBILBANK','ANT_INNLOGGINGER_NETTBANK']].sum()
df_grouped_1.tail()
```

```python
df_grouped_2=df.groupby('KUNDE_ID')[['ALDER']].mean()
df_grouped_2.tail()
```

```python
df_grouped_1.shape
```

```python
df_grouped_2.shape
```

```python
# merge two tables
data = pd.merge(df_grouped_1, df_grouped_2, on=['KUNDE_ID'] , how='inner')
data.sort_values(by=['KUNDE_ID'], ascending=[True], inplace=True)

# Rounding the age since 'mean' produces float
data[["ALDER"]] = data[["ALDER"]].astype("int")
data
```

```python
data.shape
```

### Data normalization

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1

```python
# replace (original value) by (original value)/(maximum value)
data['INSKUDD'] = data['INSKUDD']/data['INSKUDD'].max()

data['UTLAN'] = data['UTLAN']/data['UTLAN'].max()

data['LONNSOMHET_PER_MND'] = data['LONNSOMHET_PER_MND']/data['LONNSOMHET_PER_MND'].max()
```

### Binning data in accorance with 'Lnnsomhet'

```python
# Convert data to correct format
#df["LONNSOMHET_PER_MND"]=df["LONNSOMHET_PER_MND"].astype(int, copy=True)
```

```python
bins = np.linspace(min(df["LONNSOMHET_PER_MND"]), max(df["LONNSOMHET_PER_MND"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

 We apply the functmion "cut" the determine what each value of "LONNSOMHET_PER_MND" belongs to. 

```python
df['LONNSOMHET_PER_MND-binned'] = pd.cut(df['LONNSOMHET_PER_MND'], bins, labels=group_names, include_lowest=True )
df[['LONNSOMHET_PER_MND','LONNSOMHET_PER_MND-binned']].head(20)
```

```python
# The number of cutsomers in each bin
df["LONNSOMHET_PER_MND-binned"].value_counts()
```

Lets plot the distribution of each bin.

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["LONNSOMHET_PER_MND-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("LONNSOMHET_PER_MND")
plt.pyplot.ylabel("count")
plt.pyplot.title("lonnsomhet bins")
```
